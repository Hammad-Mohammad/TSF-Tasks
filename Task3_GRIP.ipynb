# Exploratory Data Analysis

```python
#Visualising the dataset s whole using the pairplot on the basis of categories
sb.pairplot(store_df, hue = 'Category')
```

```python
#Finding the paiwise correlation between the attributes and visualing them using heatmap
cor = store_df.corr()
cor
#Visualising using Heatmap
plt.subplots(figsize = (10,10))
sb.heatmap(cor, annot = True)
```

```python
#Visualizing the categories of items sold using pie chart 
store_df.Category.value_counts().plot( kind = 'pie' , autopct= '%1.01f%%' , explode = (0.05, 0.05, 0.05))
```

```python
#Plotting the overall profit and overall sales of different categories of items 
store_df.groupby('Category')['Sales', 'Profit'].agg(['sum']).plot.bar()
```

```python
#Calculating the top selling category of items 
transform_s = store_df.groupby('Category').Sales.sum()
#calculating the top profit making category of items
transform_p = store_df.groupby('Category').Profit.sum()

#Visualising to understand it concisely
transform_s.plot( kind = 'bar' , fontsize = 10 , legend = True)
transform_p.plot( kind = 'bar' , fontsize = 10 , color = 'crimson' ,legend = True)
plt.title('Top Category Sales VS Profit')
```

```python
#Calculating the top regions in term of sales made
transform_s = store_df.groupby('Region').Sales.sum()
#Calculating the top regions in term of profit generated
transform_p = store_df.groupby('Region').Profit.sum()

#Visualising to understand concisely
transform_s.plot( kind = 'bar' , fontsize = 10, legend = True )
transform_p.plot( kind = 'bar' , fontsize = 10, color = 'red', legend = True )
```

```python
#Calculating the sales made in states
transform_s = store_df.groupby('State').Sales.sum()
#Calculating the profit made in states
transform_p = store_df.groupby('State').Profit.sum()

#Visualising to understand concisely
plt.subplots(figsize = (20,5))
transform_s.plot( kind = 'bar' , fontsize = 12)
transform_p.plot( kind = 'bar' , fontsize = 12 , color = 'orange')
plt.show()
```

```python
#Calculating the sales made in product subcategory
transform_s = store_df.groupby('Sub-Category').Sales.sum()
#Calculating the sales made in product subacategory
transform_p = store_df.groupby('Sub-Category').Profit.sum()

#Visualising to understand concisely
plt.subplots(figsize = (15,6))
transform_s.plot( kind = 'bar' , fontsize = 10)
transform_p.plot( kind = 'bar' , fontsize = 10 , color = 'red')
plt.show()
```

```python
#plotting the profit made in product subcategory horizontally
plt.subplots(figsize = (15,6))
transform_p.plot( kind = 'barh' , fontsize = 10 , color = 'red')
plt.show()
```

```python
#Plotting the discount and profit
plt.subplots(figsize = (6,6))
sb.lineplot(x = store_df['Discount'], y = store_df['Profit'])
```

```python
#Visualising the share of Profit over different regions using pie charts
store_df.groupby('Region').sum().plot(kind = 'pie', y = 'Profit',autopct='%1.01f%%', explode = (0.025, 0.025, 0.025, 0.025), legend= False)
plt.show()
```

```python
#scatter plot on interdependency of sales and profit with respect to discount
store_df.plot(kind = 'scatter', figsize = (10,6), x = 'Sales', y = 'Profit', c = 'Discount', cmap= 'viridis')
```
